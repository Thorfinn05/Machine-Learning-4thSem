```python
import csv
import pandas as pd
```

```python
data = pd.read_csv('data2.csv')
print(data)
training_data = data.iloc[:, [0,1]].values.tolist()
labels = data.iloc[:, 2].astype(int).tolist()
print(training_data)
print(labels)
```

```python
test_data = [1,4]
k=3
print(test_data)
print(k)
```

```python
distances = []
for i in range(len(training_data)):
    d = ((training_data[i][0] - test_data[0])**2 + (training_data[i][1] - test_data[1])**2)**0.5
distances.append([d, labels[i]])
print(distances)
```

```python
for i in range(len(distances)):
    for j in range(i+1, len(distances)):
        if(distances[i] > distances[j]):
            distances[i], distances[j] = distances[j], distances[i]
print("Sorted distance list: ", distances)
```

```python
neighbours = []
for i in range(k):
    neighbours.append(distances[i][1])
print("The first k nearest neighbour in increasing order: ", neighbours)
```
