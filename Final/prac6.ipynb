```python
import csv
```

```python
data = []
with open("D:\\Github\\Machine-Learning-4thSem\\Final\\trans.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)
    for row in reader:
        data.append(row[1:])
print(data) 
```

```python
min_support =2
min_confidence = 0.6
```

```python
itemsets = []
for x in data:
    for product in x:
        if product not in itemsets:
            itemsets.append(product)
print('All unique products: ', itemsets)
```

```python
def count_freq(itemsets):
    count = 0
    for items in data:
        present = True
        for product in itemsets:
            if product not in items:
                present = False
                break
        if present == True:
            count += 1
    return count
```

```python
from itertools import combinations
count_comb_freq = {}
for k in range(1, len(itemsets)+1):
    c = combinations(itemsets, k)
    for x in c:
        if(count_freq(x) >= min_support):
            count_comb_freq[x] = count_freq(x)
print(count_comb_freq)
```

```python
for comb in count_comb_freq:
    print(comb, " were bought ", count_comb_freq[comb], ' times.')
```

```python
for items in count_comb_freq:
    if(len(items) >= 2):
        for k in range(1, len(items)):
            for left in combinations(items, k):
                right = tuple(sorted(set(items) - set(left)))
                if left in count_comb_freq and right in count_comb_freq:
                    confidence = count_comb_freq[items]/count_comb_freq[left]
                if confidence >= min_confidence:
                    print(left, " ---> ", right, ' : ', round(confidence, 4))
```
