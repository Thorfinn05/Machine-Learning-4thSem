```python
import csv
```

```python
data = []
with open("D:\\Github\\Machine-Learning-4thSem\\Final\\trans.csv", "r")as file:
    reader = csv.reader(file)
    for row in reader:
        data.append(row[1:])
print(data)
```

```python
min_support = 2
min_confidence = 0.6
```

```python
items = []
for x in data:
    for product in x:
        if product not in items:
            items.append(product)
print("All unique products in transaction dataset: ", items)
```

```python
def count_freq(itemsets):
    count = 0
    for x in data:
        for product in itemsets:
            present = True
            if product not in x:
                present = False
                break
        if present == True:
            count += 1
    return count
```

```python
from itertools import combinations
count_comb_freq = {}
for k in range(1, len(items)+1):
    c = (combinations(items, k))
    for x in c:
        print(x," :", count_freq(x))
        if(count_freq(x) >= min_support):
            count_comb_freq[x] = count_freq(x)
print(count_comb_freq)
```

```python
for comb in count_comb_freq:
    if(count_comb_freq[comb] == 1):
        print(f"{comb} was bought {count_comb_freq[comb]} times.")
    else:
        print(f"{comb} were bought together {count_comb_freq[comb]} times.")
```

```python
for items in count_comb_freq:
    if(len(items) >= 2):
        for k in range(1, len(items)):
            for left in combinations(items, k):
                right = tuple(sorted(set(items) - set(left)))
                if left in count_comb_freq and right in count_comb_freq:
                    confidence = count_comb_freq[items]/count_comb_freq[left]
                    if confidence >= min_confidence:
                        print(left, " ---> ", right, " : ", confidence)
```
