```python
import pandas as pd
import math
```

```python
data = pd.read_csv("D:\Github\Machine-Learning-4thSem\Final\Weather-D.csv")
print(data)
```

```python
Sunny = len(data[data["Outlook"] == "Sunny"])
print(Sunny)
Sunny_Y = len(data[(data["Outlook"] == "Sunny") & (data["Play Football"] == "Yes")])
print(Sunny_Y)
```

```python
def entropy(pos, neg):
    entropy = 0
    total = pos+neg
    if total == 0 or pos == 0 or neg == 0:
        return 0
    p_pos = pos/total
    p_neg = neg/total
    return -p_pos * math.log2(p_pos) - p_neg * math.log2(p_neg)
```

```python
pos = len(data[data["Play Football"] == "Yes"])
neg = len(data[data["Play Football"] == "No"])
total = pos + neg
total_entropy = entropy(pos, neg)
print("Total Data: ", total)
print('Total yes: ', pos)
print("Total No: ", neg)
print("Toal Entropy: ", total_entropy)
```

```python
def calc_gain(attribute):
    categories = data[attribute].unique()
    # print(categories) #For Outlook it would be ['Sunny' 'Overcast' 'Rainy']
    weighted_entropy = 0
    for cat in categories:
        subset = data[data[attribute] == cat]
        pos = len(subset[subset["Play Football"] == "Yes"])
        neg = len(subset[subset['Play Football'] == "No"])
        ent = entropy(pos, neg)
        weight = len(subset)/total
        weighted_entropy += weight * ent
    gain = total_entropy - weighted_entropy
    return round(gain, 4)
    print(subset)


gain_outlook = calc_gain("Outlook")
print(gain_outlook)
gain_humidity = calc_gain("Humidity")
gain = {
    "Outlook" : gain_outlook,
    "Humidity" : gain_humidity
}
print(gain)
root_node = max(gain, key=gain.get)
print(root_node)
```
